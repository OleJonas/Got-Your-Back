### En måte å skille ut sensorer til nytt datasett basert på hvor mange sensorer vi ønsker bruke

```python
def split_dataset(dataframe, number_of_sensors=3):
    assert (number_of_sensors > 0 and number_of_sensors <= 3), "Feil antall sensorer!"
    if (number_of_sensors == 1):
        return dataframe[(dataframe['SensorId'] == 1)]
    elif(number_of_sensors == 2):
        return dataframe[(dataframe['SensorId'] == 1) | (dataframe['SensorId'] == 2)]
    else:
        return dataframe[(dataframe['SensorId'] == 1) | (dataframe['SensorId'] == 2) | (dataframe['SensorId'] == 3)]

df_test = split_dataset(df, 1)
df_test.iloc[:2000, :4]
```
